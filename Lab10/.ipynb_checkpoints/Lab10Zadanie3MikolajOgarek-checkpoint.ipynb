```python
#Zadanie 3
from math import floor, ceil

def podziel(d1,g1,d2,g2):
    c = [-23,-17]
    A_ub = [[4,3],[1,1],[-1,0],[1,0],[0,-1],[0,1]]
    b_ub = [190,55,-d1,g1,-d2,g2]


    x1_bounds = (0,None)
    x2_bounds = (0,None)


    from scipy.optimize import linprog
    res = linprog(c, A_ub=A_ub, b_ub=b_ub,bounds=[x1_bounds, x2_bounds])
    
    
    if(res.success==False):
        return (0)
    elif(res.x[0]!=round(res.x[0])):
        a = podziel(ceil(res.x[0]),g1,d2,g2)
        b = podziel(d1,floor(res.x[0]),d2,g2)
        
        if (a>b):
            c = [-23,-17]
            A_ub = [[4,3],[1,1],[-1,0],[1,0],[0,-1],[0,1]]
            b_ub = [190,55,-ceil(res.x[0]),g1,-d2,g2]


            x1_bounds = (0,None)
            x2_bounds = (0,None)

            res = linprog(c, A_ub=A_ub, b_ub=b_ub,bounds=[x1_bounds, x2_bounds])
            
            tab.append((res.x[0], res.x[1], -res.fun))
            
            return a
        
        else:
            c = [-23,-17]
            A_ub = [[4,3],[1,1],[-1,0],[1,0],[0,-1],[0,1]]
            b_ub = [190,55,-d1,floor(res.x[0]),-d2,g2]


            x1_bounds = (0,None)
            x2_bounds = (0,None)

            res = linprog(c, A_ub=A_ub, b_ub=b_ub,bounds=[x1_bounds, x2_bounds])
            
            tab.append((res.x[0], res.x[1], -res.fun))
            
            return b
    
    elif(res.x[1]!=round(res.x[1])):
        
        return(max(podziel(d1,g1,ceil(res.x[1]),g2),podziel(d1,g1,d2,floor(res.x[1]))))
        a = podziel(d1,g1,ceil(res.x[1]),g2)
        b = podziel(d1,g1,d2,floor(res.x[1]))
        
        if (a>b):
            c = [-23,-17]
            A_ub = [[4,3],[1,1],[-1,0],[1,0],[0,-1],[0,1]]
            b_ub = [190,55,-d1,g1,-ceil(res.x[1]),g2]


            x1_bounds = (0,None)
            x2_bounds = (0,None)

            res = linprog(c, A_ub=A_ub, b_ub=b_ub,bounds=[x1_bounds, x2_bounds])
            
            tab.append((res.x[0], res.x[1], -res.fun))
            
            return a
        
        else:
            c = [-23,-17]
            A_ub = [[4,3],[1,1],[-1,0],[1,0],[0,-1],[0,1]]
            b_ub = [190,55,-d1,g1,-d2,podziel(d1,g1,d2,floor(res.x[1]))]


            x1_bounds = (0,None)
            x2_bounds = (0,None)

            res = linprog(c, A_ub=A_ub, b_ub=b_ub,bounds=[x1_bounds, x2_bounds])
            
            tab.append((res.x[0], res.x[1], -res.fun))
            
            return b
    
    else:
        tab.append((res.x[0], res.x[1]))
        return ((-res.fun))
    

podziel(-1000,1000,-1000,1000)
print(tab[1])   

```
